```python
import pandas as pd

# Assuming df is your DataFrame

# Initialize lists to store data
uids = []
totals = []

columns_list = ['LINEAR ALGEBRA & DIFFERENTIAL CALCULUS',
               'ENGINEERING CHEMISTRY',
               'BASIC ELECTRICAL TECHNOLOGY',
               'ENGINEERING GRAPHICS',
               'COMPUTATIONAL MATHEMATICS LAB - I',
               'ENGINEERING CHEMISTRY LAB',
               'BASIC ELECTRICAL TECHNOLOGY LAB',
               'ENGINEERING GRAPHICS LAB',
               'PYTHON PROGRAMMING LAB',
               'WORKSHOP I',
               'COMMUNICATION SKILLS',
               'CONSTITUTION OF INDIA (NON-CREDIT)',
               'ESSENCE OF INDIAN TRADITIONAL KNOWLEDGE (NON-CREDIT)']

# Initialize dictionary to store TOT values for each UID
tot_dict = {}

# Iterate over the rows
for index, row in df.iterrows():
    # Check if the UID starts with "2019"
    if str(row[0]).startswith('2019'):
        # Append UID
        uids.append(str(row[0]))
        # Iterate over the columns
        for column_name, value in row.items():
            # Check if the column name is 'TOT'
            if column_name == 'TOT':
                # Add TOT value to the dictionary
                if str(row[0]) not in tot_dict:
                    tot_dict[str(row[0])] = []
                tot_dict[str(row[0])].append(value)
                print(tot_dict)

# Convert the dictionary to a list of dictionaries
for uid, tot_values in tot_dict.items():
    for tot_value in tot_values:
        totals.append({'UID': uid, 'TOT': tot_value})

# Create a DataFrame from the list of dictionaries
df_new = pd.DataFrame(totals)

df_new
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Assuming tot_dict is created earlier

# Define the list of TOT column names
linear_algebra_differential_calculus = []
engineering_chemistry = []
basic_electrical_technology = []
engineering_graphics = []
computational_mathematics_lab = []
engineering_chemistry_lab = []
basic_electrical_technology_lab = []
engineering_graphics_lab = []
python_programming_lab = []
workshop_i = []
communication_skills = []
constitution_of_india = []
indian_traditional_knowledge = []

# Initialize lists to store data
uids = []
names= []

# Iterate over the rows
for index, row in df.iterrows():
    # Check if the UID starts with "2019"
    if str(row[0]).startswith('2019'):
        # Append UID
        uid = str(row[0])
        uids.append(uid)
        name_val = str(df.iloc[index, 0]).strip('/')  # Remove leading '/'
        # print(name_val)
        names.append(name_val)

        for column_name, value in row.items():
            # Check if the column name is 'TOT'
            if column_name == 'TOT':
                # Add TOT value to the dictionary
                if str(row[0]) not in tot_dict:
                    tot_dict[str(row[0])] = []
                tot_dict[str(row[0])].append(value)

# Iterate over tot_dict and append values to the lists
for uid, tot_values in tot_dict.items():
    if uid.startswith('2019'):
        linear_algebra_differential_calculus.append(tot_values[0])
        engineering_chemistry.append(tot_values[1])
        basic_electrical_technology.append(tot_values[2])
        engineering_graphics.append(tot_values[3])
        computational_mathematics_lab.append(tot_values[4])
        engineering_chemistry_lab.append(tot_values[5])
        basic_electrical_technology_lab.append(tot_values[6])
        engineering_graphics_lab.append(tot_values[7])
        python_programming_lab.append(tot_values[8])
        workshop_i.append(tot_values[9])
        communication_skills.append(tot_values[10])
        constitution_of_india.append(tot_values[11])
        indian_traditional_knowledge.append(tot_values[12])

data = {
    'UID': uids,
    'Name': names,
    'Linear_Algebra_Differential_Calculus': linear_algebra_differential_calculus,
    'Engineering_Chemistry': engineering_chemistry,
    'Basic_Electrical_Technology': basic_electrical_technology,
    'Engineering_Graphics': engineering_graphics,
    'Computational_Mathematics_Lab': computational_mathematics_lab,
    'Engineering_Chemistry_Lab': engineering_chemistry_lab,
    'Basic_Electrical_Technology_Lab': basic_electrical_technology_lab,
    'Engineering_Graphics_Lab': engineering_graphics_lab,
    'Python_Programming_Lab': python_programming_lab,
    'Workshop_I': workshop_i,
    'Communication_Skills': communication_skills,
    'Constitution_of_India': constitution_of_india,
    'Indian_Traditional_Knowledge': indian_traditional_knowledge
}

# Create DataFrame
df_new = pd.DataFrame(data)

# Display DataFrame
df_new
```
