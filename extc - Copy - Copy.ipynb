```python
import os
import pandas as pd
```

```python
df = pd.read_excel('./SE/20-21 DSY EXTC SEM IV GR 2 Gazette MAY 2021.xls')
df.columns = df.iloc[4]
df = df.drop(4)
df.head(15)
```

```python
# df.iloc[3]
# Assuming df is your DataFrame
non_nan_values = [value for value in df.iloc[3] if pd.notna(value)]
non_nan_values
```

```python
len(non_nan_values)
```

```python

['PROBABILITY AND STOCHASTIC PROCESSES',
 'LINEAR ALGEBRA',
 'FOUNDATION OF MATHEMATICS - II',
 'ANALOG CIRCUITS',
 'MICROCONTROLLERS',
 'SIGNALS AND SYSTEMS',
 'MINI PROJECT - I ',
 'HSS - II',
 'CONSTITUTION OF INDIA']

probability_and_stochastic_processes = []
linear_algebra = []
foundation_of_mathematics_ii =[]
analog_circuits = []
microcontrollers = []
signals_and_systems = []
mini_project_i = []
hss_ii = []
constitution_of_india = []

# Initialize lists to store data
uids = []
names= []

tot_dict = {}

# Iterate over the rows
for index, row in df.iterrows():
    # Check if the UID starts with "2019"
    if str(row[0]).startswith('20'):
        # Append UID
        uid = str(row[0])
        uids.append(uid)
        name_val = str(df.iloc[index, 0]).strip('/')  # Remove leading '/'
        # print(name_val)
        names.append(name_val)

        for column_name, value in row.items():
            # Check if the column name is 'TOT'
            if column_name == 'TOT':
                # Add TOT value to the dictionary
                if str(row[0]) not in tot_dict:
                    tot_dict[str(row[0])] = []
                tot_dict[str(row[0])].append(value)

# Iterate over tot_dict and append values to the lists
for uid, tot_values in tot_dict.items():
    if uid.startswith('20'):
        probability_and_stochastic_processes.append(tot_values[0])
        linear_algebra.append(tot_values[1])
        foundation_of_mathematics_ii.append(tot_values[2])
        analog_circuits.append(tot_values[3])
        microcontrollers.append(tot_values[4])
        signals_and_systems.append(tot_values[5])
        mini_project_i.append(tot_values[6])
        hss_ii.append(tot_values[7])
        constitution_of_india.append(tot_values[8])

data = {
    'UID': uids,
    'Name': names,
    'Probability_And_Stochastic_Processes': probability_and_stochastic_processes,
    'Linear_Algebra': linear_algebra,
    'Foundation_Of_Mathematics_II': foundation_of_mathematics_ii,
    'Analog_Circuits': analog_circuits,
    'Microcontrollers': microcontrollers,
    'Signals_And_Systems': signals_and_systems,
    'Mini_Project_I': mini_project_i,
    'HSS_II': hss_ii,
    'Constitution_Of_India': constitution_of_india
}

# Create DataFrame
df_new = pd.DataFrame(data)

# Display DataFrame
df_new
```

```python
df_new.to_csv('Sem IV dsy Res.csv', index=False)
```
