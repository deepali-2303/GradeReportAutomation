```python
# Dataframes - regular, dsy, pastYear

import pandas as pd

df = pd.read_csv('merged_df2.csv')


df['UID'] = df['UID'].astype(str)


regular_df = df[df['UID'].str.startswith('2019')]
dsy_df = df[df['UID'].str.startswith('2020')]
prev_df = df[df['UID'].str.startswith('2018')]

# regular_df
prev_df
```

```python
## First Year - regular students 


FEcount_F = 0
FEcount_P = 0

FEbacklog_list = []
FEregular_list = []


for index, row in regular_df.iterrows():
    
    result1 = row['Results_1']
    result2 = row['Results_2']
    
   
    if (result1 == 'F' or result2 == 'F'):
        FEcount_F += 1
        FEbacklog_list.append(row['UID'])
    elif (result1 == 'P' and result2 == 'P'):
        FEcount_P += 1
        FEregular_list.append(row['UID'])

# Display the counts
print("Number of F (FE):", FEcount_F)
print("Number of P (FE):", FEcount_P)

# Display the backlog and regular lists
print("Backlog List (FE):", FEbacklog_list)
print("Regular List (FE):", FEregular_list)
```

```python
## Second Year - regular students

SEreg_count_F = 0
SEreg_count_P = 0
SEreg_count_N = 0


SEbacklog_list = []
SEregular_list = []
SEdsy_empty_list = []



for index, row in regular_df.iterrows():
    
    result3 = row['Results_3']
    result4 = row['Results_4']
    
   
    if (result3 == 'F' or result4 == 'F'):
        SEreg_count_F += 1
        SEbacklog_list.append(row['UID'])
    elif (result3 == 'P' and result4 == 'P'):
        SEreg_count_P += 1
        SEregular_list.append(row['UID'])
    else:
        SEreg_count_N +=1
        SEdsy_empty_list.append(row['UID'])    

# Display the counts
print("Number of F (SE-reg):", SEreg_count_F)
print("Number of P (SE-reg):", SEreg_count_P)
print("Number of NaN (SE-reg):", SEreg_count_N)

# Display the backlog and regular lists
print("Backlog List (SE-reg):", SEbacklog_list)
print("Regular List (SE-reg):", SEregular_list)
print("Empty List (SE-reg):", SEdsy_empty_list)
```
