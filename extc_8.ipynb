```python
import pandas as pd
```

```python
df_inhouse = pd.read_excel('./BE/EXTC SEM VIII MAY 2023.xlsx', sheet_name='Inhouse')
df_inhouse = df_inhouse.iloc[:, 1:]
df_inhouse.columns = df_inhouse.iloc[3]
df_inhouse = df_inhouse.drop(3)
df_inhouse.head()
```

```python
uids = []
names = []
sgpa = []
results = []

# Iterate over the rows
for index, row in df_inhouse.iterrows():
    # Check if the string starts with "2019"
    if str(row[0]).startswith('20'):
        # Append UID and Name
        # print(str(row[0]))
        uids.append(str(row[0]))
        name_val = str(df_inhouse.iloc[index - 1, 1]).strip('/')  # Remove leading '/'
        names.append(name_val)
        # Append SGPA and results
        sgpa_val = str(row['SEM VIII SGPA'])  # Assuming 'SGPA' is the column name for SGPA
        try:
            sgpa.append(float(sgpa_val))
        except ValueError:
            sgpa.append(None)
        results_val = str(row['Remark'])  # Assuming 'Results' is the column name for results
        # print(results_val)
        results.append(results_val)

# Create a DataFrame from the lists
df_new_inhouse = pd.DataFrame({'UID': uids, 'Name': names, 'SGPA': sgpa, 'Results': results})

df_new_inhouse
```
